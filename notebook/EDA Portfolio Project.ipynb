```python
# Clean Price column
def clean_price(price_str):
    if pd.isna(price_str):
        return np.nan

    # Convert to string and remove 'PKR', '\n' and commas
    price_str = str(price_str).replace('PKR', '').replace('\n', '').replace(',', '').strip()

    # Handle 'Crore' and 'Lakh'
    if 'Crore' in price_str:
        value = float(price_str.replace('Crore', '').strip()) * 10_000_000
    elif 'Lakh' in price_str:
        value = float(price_str.replace('Lakh', '').strip()) * 100_000
    else:
        try:
            value = float(price_str)
        except ValueError:
            value = np.nan
    return value
```

```python
# Apply the cleaning function to the 'Price' column
property['Price'] = property['Price'].apply(clean_price)
```

```python
# Standardize city names using Fuzzy Matching
# fuzzywuzzy library if not already installed
!pip install fuzzywuzzy
!pip install python-Levenshtein

import Levenshtein
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
# unique city names
unique_cities = property['City'].dropna().unique().tolist()
print(f"Number of unique cities before standardization: {len(unique_cities)}")
print("Sample of unique cities:")
display(unique_cities[:20])
```

```python
# standard list of cities
# top 10 most frequent cities as 'standard'
standard_cities_base = property['City'].value_counts().head(10).index.tolist()

standard_cities_list = list(set(standard_cities_base + ['Karachi', 'Lahore', 'Islamabad', 'Rawalpindi', 'Faisalabad', 'Multan', 'Hyderabad', 'Peshawar', 'Quetta', 'Sialkot']))
standard_cities_list = sorted(list(set(standard_cities_list))) # Remove duplicates and sort

print("Standard city names for fuzzy matching:")
display(standard_cities_list)
```

```python
def standardize_city_name(city_name, standard_list, threshold=80):
    if pd.isna(city_name):
        return np.nan

    city_name = str(city_name).strip()
    if city_name in standard_list:
        return city_name

    match = process.extractOne(city_name, standard_list, scorer=fuzz.token_set_ratio)

    if match and match[1] >= threshold:
        return match[0]
    else:
        return city_name
```

```python
# Apply to the 'City' column
property['City_Standardized'] = property['City'].apply(lambda x: standardize_city_name(x, standard_cities_list))
```

```python
# Display the old and new City columns and check
display(property[['City', 'City_Standardized']].head(20))

print(f"Number of unique cities after standardization: {len(property['City_Standardized'].dropna().unique())}")
print("Sample of unique standardized cities:")
display(property['City_Standardized'].dropna().unique().tolist()[:20])
```

```python
# Detect and correct inconsistent property types

unique_property_types = property['Type'].value_counts(dropna=False)
print("Unique Property Types and their counts before standardization:")
display(unique_property_types)
```

```python
# Also display just the unique types for easier review
print("\nUnique Property Types (list):")
display(property['Type'].dropna().unique().tolist())
```

```python
# standard property types
standard_property_types = [
    'House', 'Flat', 'Upper Portion', 'Lower Portion', 'Penthouse',
 'Farm House','Room'
]

def standardize_property_type(property_type_name, standard_list, threshold=80):
    if pd.isna(property_type_name):
        return np.nan

    property_type_name = str(property_type_name).strip()
    if property_type_name in standard_list:
        return property_type_name

    match = process.extractOne(property_type_name, standard_list, scorer=fuzz.ratio)

    if match and match[1] >= threshold:
        return match[0]
    else:
        return property_type_name
```

```python
# Apply the standardization function to the 'Type' column
property['Type_Standardized'] = property['Type'].apply(lambda x: standardize_property_type(x, standard_property_types))
```

```python
# Display the old and new Type columns and check unique values after standardization
display(property[['Type', 'Type_Standardized']].head(20))

print(f"\nNumber of unique property types after standardization: {len(property['Type_Standardized'].dropna().unique())}")
print("Unique Standardized Property Types (list):")
display(property['Type_Standardized'].dropna().unique().tolist())
```

```python
# Remove outliers using IQR method or Z-score method

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Keep rows that are within the bounds
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    # Calculate number of outliers removed
    num_outliers = len(df) - len(df_filtered)
    print(f"Removed {num_outliers} outliers from '{column}' using IQR method.")

    return df_filtered
```

```python
# numerical columns for outlier removal
numerical_cols_for_outliers = ['Price', 'Built in year', 'Kitchens', 'Area']

initial_shape = property.shape
print(f"Initial DataFrame shape: {initial_shape}")

for col in numerical_cols_for_outliers:

    if pd.api.types.is_numeric_dtype(property[col]):
        property = remove_outliers_iqr(property, col)
    else:
        print(f"Skipping non-numeric column: {col}")
```

```python
final_shape = property.shape
print(f"Final DataFrame shape after outlier removal: {final_shape}")
print(f"Total rows removed: {initial_shape[0] - final_shape[0]}")

display(property.head())
```
